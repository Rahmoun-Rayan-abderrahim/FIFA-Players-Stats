```python
import kagglehub
import shutil
import os

# Download original dataset to cache
path = kagglehub.dataset_download("stefanoleone992/fifa-21-complete-player-dataset")

# Path of your notebook directory (current directory)
target_dir = os.getcwd()

# Copy ALL dataset files to the notebook folder
for file in os.listdir(path):
    shutil.copy(os.path.join(path, file), target_dir)

print("Dataset copied to:", target_dir)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
#this is to merge all the files in one i think 

folder = "data"
csv_files = [f for f in os.listdir(folder) if f.endswith(".csv")]
csv_files
```

```python
#Loading all CSVs into a single DataFrame 

df_list = []

for file in csv_files:
    temp_df = pd.read_csv(os.path.join(folder, file))
    temp_df['year'] = int(file.split('_')[1].split('.')[0])  # adding a new columns that will endicate the year edition the player comes from 
    df_list.append(temp_df)

df = pd.concat(df_list, ignore_index=True)
df.head()
```

```python
#realy usfull func for nowing the columns names 

print(df.columns)

#info of the data set 
df.info()
```

```python
#top 10 nationality with the most number of players over all years 

df['nationality'].value_counts().head(10)
```

```python

# Top 10 players
df[['long_name', 'overall', 'club_name', 'year']].sort_values(by='overall', ascending=False).head(10)

```

```python
# Top player per year using groupby
top_per_year = df.sort_values(['year', 'overall'], ascending=[True, False]).groupby('year').head(1)

top_per_year[['year', 'short_name', 'overall', 'club_name']]
```

```python
pos_count = df.groupby(['year','player_positions']).size().unstack(fill_value=0)
pos_count.plot(kind='bar', stacked=True, figsize=(12,6))
plt.title("Number of Players per Position by FIFA Year")
plt.ylabel("Number of Players")
plt.show()
```

```python
#Best club acroos all yars Editions

df['club_name'].value_counts().head(15).plot(kind='barh', figsize=(10,6))
plt.title("Top Clubs Across FIFA Editions")
plt.show()
```
